```python
from statistics import mean

from cheapest_insertion import CheapestInsertion
from instance import Instance
from utils import draw_solution
from random import seed

import seaborn as sns
import numpy as np


sns.set()
np.random.seed(0)
seed(0)

for instance_name in ['kroA100', 'kroB100']:
    instance = Instance(name=instance_name)
    solve_strategy: CheapestInsertion = CheapestInsertion(instance=instance)
    solve_strategy.run(run_times=10)

    best_solution = min(solve_strategy.solutions, key=lambda x: x[1])
    worst_solution = max(solve_strategy.solutions, key=lambda x: x[1])
    mean_solution = mean(x[1] for x in solve_strategy.solutions)

    print(instance_name, best_solution[1], worst_solution[1], mean_solution)

    draw_solution(
        instance=instance,
        solution=solve_strategy.solution,
        title='Cheapest_insertion{}'.format(instance_name)
    )
```

```python
from statistics import mean

from cheapest_insertion import CheapestInsertion
from instance import Instance
from utils import draw_solution

for instance_name in ['kroA100', 'kroB100']:
    instance = Instance(name=instance_name)
    solve_strategy: CheapestInsertion = CheapestInsertion(instance=instance,  regret=1)
    solve_strategy.run(run_times=10)

    best_solution = min(solve_strategy.solutions, key=lambda x: x[1])
    worst_solution = max(solve_strategy.solutions, key=lambda x: x[1])
    mean_solution = mean(x[1] for x in solve_strategy.solutions)

    print(instance_name, best_solution[1], worst_solution[1], mean_solution)

    draw_solution(
        instance=instance,
        solution=solve_strategy.solution,
        title='Cheapest Insertion with regret{}'.format(instance_name)
    )
```
