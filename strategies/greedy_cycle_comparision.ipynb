```python
from collections import defaultdict
from random import seed

import seaborn as sns
import numpy as np
from tqdm import tqdm_notebook
from cheapest_insertion import CheapestInsertion
from instance import Instance
from utils import draw_solution

sns.set()
np.random.seed(0)
seed(0)

def get(percentage):
        instance = Instance(name=instance_name)
        solve_strategy: CheapestInsertion = CheapestInsertion(
            instance=instance,
            regret=regret,
            path_length_percentage=percentage,
        )
        solve_strategy.run(run_times=100)
        return solve_strategy.solution_cost

results = defaultdict(list)
for p in tqdm_notebook(range(50,101,5)):
    for instance_name in ['kroA100', 'kroB100']:
        for regret in [0, 1]:
                results[instance_name+str(regret)].append(get(p))
results
```

```python
results['x'] = list(range(50,101,5))
```
